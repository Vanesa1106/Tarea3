## Vinos

```python
#Importar libreria para manipular datos
import pandas as pd
datos_vinos = pd.read_csv('Vinos/winequality-red.csv')
```

```python
datos_vinos
```

```python
filas_vacias = datos_vinos.isna().all(axis=1)
print("Filas completamente vacías:", filas_vacias.sum())
```
