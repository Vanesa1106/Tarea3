### Selección de caracteristicas
Procedemos a verificar la correlación de cada variable del dataset con la calidad. La correlación de Pearson nos dice qué tan fuerte es la relación entre cada variable y la calidad del vino.


```python
import seaborn as sns
import matplotlib.pyplot as plt

# Calcular la correlación
corr = datos_vinos.corr()

# Mostrar la correlación con "quality" en orden descendente
print(corr["quality"].sort_values(ascending=False))

# Mapa de calor para visualizar correlaciones
plt.figure(figsize=(10, 6))
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.show()
```

Como resultado tenemos que las características más prometedoras hasta ahora son:

1. *alcohol* tiene la mayor correlación con quality (0.47). 

2. *sulphates* también tiene una correlación moderada (0.25). 

3. *citric acid* muestra una correlación positiva (0.22).

4. *volatile acidity* tiene una correlación negativa fuerte (-0.39), lo que sugiere que valores altos de acidez volátil afectan negativamente la calidad. 

5. *total sulfur dioxide* tiene una correlación negativa (-0.20), lo que indica que un alto contenido de este compuesto reduce la calidad del vino.



```python
from sklearn.feature_selection import SelectKBest, f_regression
import pandas as pd
X = datos_vinos.drop(columns=["quality"])  # Eliminamos 'quality' porque es la variable objetivo
y = datos_vinos["quality"]
selector = SelectKBest(score_func=f_regression, k=5)  # k=5 selecciona las 5 mejores características
X_selected = selector.fit_transform(X, y)
selected_features = X.columns[selector.get_support()]
print("Características más relevantes:", selected_features)
```

#### Procedemos a seleccionar el set de datos de entrenamoiento y prueba

```python
from sklearn.model_selection import train_test_split

# Filtramos las columnas seleccionadas
X_selected_df = datos_vinos[selected_features]  

# Dividimos en 80% entrenamiento y 20% prueba
X_train, X_test, y_train, y_test = train_test_split(X_selected_df, y, test_size=0.2, random_state=42)

print("Tamaño del conjunto de entrenamiento:", X_train.shape)
print("Tamaño del conjunto de prueba:", X_test.shape)
```

#### Procedemos a entrenar el modelo de regresión líneal

```python
from sklearn.linear_model import LinearRegression

# Inicializamos el modelo
modelo = LinearRegression()

# Entrenamos con el conjunto de entrenamiento
modelo.fit(X_train, y_train)

# Mostramos los coeficientes del modelo
print("Coeficientes del modelo:", modelo.coef_)
print("Intercepto del modelo:", modelo.intercept_)
```

#### Evalución de modelo la metrica de precekisopk

```python
from sklearn.metrics import mean_squared_error, r2_score

# Hacemos predicciones con el conjunto de prueba
y_pred = modelo.predict(X_test)

# Evaluamos el modelo
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("Error cuadrático medio (MSE):", mse)
print("Coeficiente de determinación (R²):", r2)
```

#### Evaluacion con recall y F1-score

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import recall_score, f1_score

# 1️⃣ Convertimos la variable objetivo 'quality' en clases binarias
y_clasificacion = (datos_vinos['quality'] >= 6).astype(int)  # 1: Bueno, 0: Malo

# 2️⃣ Mantenemos las características seleccionadas previamente
X_selected_df = datos_vinos[selected_features]

# 3️⃣ Dividimos en conjunto de entrenamiento (80%) y prueba (20%)
X_train, X_test, y_train, y_test = train_test_split(X_selected_df, y_clasificacion, test_size=0.2, random_state=42)

# 4️⃣ Entrenamos un modelo de clasificación (Regresión Logística)
modelo_clasificacion = LogisticRegression()
modelo_clasificacion.fit(X_train, y_train)

# 5️⃣ Hacemos predicciones
y_pred = modelo_clasificacion.predict(X_test)

# 6️⃣ Evaluamos con Recall y F1-score
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

print("Recall:", recall)
print("F1-score:", f1)
```

### Uso de graficos: Matriz de Confusión, Curva ROC y AUC y Gráfico de Errores

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix, roc_curve, auc

# 1️⃣ Matriz de Confusión
conf_matrix = confusion_matrix(y_test, y_pred)

plt.figure(figsize=(6,4))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=['Malo', 'Bueno'], yticklabels=['Malo', 'Bueno'])
plt.xlabel("Predicción")
plt.ylabel("Real")
plt.title("Matriz de Confusión")
plt.show()

# 2️⃣ Curva ROC y AUC
y_prob = modelo_clasificacion.predict_proba(X_test)[:, 1]  # Probabilidades de la clase positiva
fpr, tpr, _ = roc_curve(y_test, y_prob)
roc_auc = auc(fpr, tpr)

plt.figure(figsize=(6,4))
plt.plot(fpr, tpr, color='blue', label=f'AUC = {roc_auc:.2f}')
plt.plot([0, 1], [0, 1], color='gray', linestyle='--')
plt.xlabel("Tasa de Falsos Positivos (FPR)")
plt.ylabel("Tasa de Verdaderos Positivos (TPR)")
plt.title("Curva ROC")
plt.legend()
plt.show()

# 3️⃣ Gráfico de Errores en la Regresión
errores = y_test - y_pred

plt.figure(figsize=(6,4))
sns.histplot(errores, bins=20, kde=True, color="red")
plt.xlabel("Error")
plt.ylabel("Frecuencia")
plt.title("Distribución de Errores del Modelo de Regresión")
plt.show()
```

#### Analisis de resultados y conclusiones
**Evaluación de Desempeño**
- Precisión, Recall y F1-score indican un rendimiento aceptable, pero con margen de mejora.

- Matriz de Confusión: El modelo logra clasificar correctamente la mayoría de los casos, pero aún presenta falsos positivos y falsos negativos, lo que puede impactar en su utilidad según el contexto.

- Curva ROC (AUC = 0.81): Indica una buena capacidad del modelo para diferenciar entre clases, aunque no es perfecto. Un AUC superior a 0.80 sugiere que el modelo tiene un buen desempeño.

- Distribución de Errores: Muestra que el modelo predice bien en la mayoría de los casos, pero comete errores extremos en algunos, lo que podría afectar la confiabilidad en ciertas situaciones.

📌 **Conclusión Final**
El modelo tiene un desempeño razonablemente bueno, con un equilibrio entre precisión y sensibilidad. Sin embargo, la presencia de errores extremos y falsas clasificaciones indica que aún hay oportunidades de mejora.
